### Se tomará una muestra de 99.000 registros

```
from sodapy import Socrata
cliente = Socrata("www.datos.gov.co", None)
datos2020 = pd.DataFrame.from_records(cliente.get('rnvb-vnyh', limit=99000)) 
datos2019 = pd.DataFrame.from_records(cliente.get('ynam-yc42', limit=99000))
```

### Diagrama de caja y bigotes del puntaje global en los 2 periodos

En la comparación de estos dos diagramas podemos decir que se presentaron más datos atípicos y mas ampliamente distribuidos en los primeros 99000 datos para los dos periodos, se presentaron puntajes más bajos así como más altos en el 2020-2 teniendo una varianza más grande, a pesar de que en el ICFES del 2020-2 hubo menos preguntas, lo que pudo haber significado una varianza menor

Además la media no cambió de gran manera

```
punt_2019 = datos2019['punt_global']
punt_2020 = datos2020['punt_global']
concat_punt_global = pd.concat([punt_2019.rename('2019'), punt_2020.rename('2020')], ignore_index=False, axis=1)

plt.boxplot(concat_punt_global, labels=['2019-2', '2020-2'])
plt.title("Diagrama de caja y bigotes del puntaje global en los 2 periodos")
```

# Cantidad Estudiantes Extranjeros 2019-2 vs 2020-2

```
estudiantes_por_pais_2019 = pd.DataFrame(datos2019['estu_nacionalidad'].value_counts())
estudiantes_por_pais_2019 = estudiantes_por_pais_2019.dropna()
estudiantes_por_pais_2020 = pd.DataFrame(datos2020['estu_nacionalidad'].value_counts())
estudiantes_por_pais_2020 = estudiantes_por_pais_2020.dropna()
```

# Balance de tipos de documento al momento de inscribirse al ICFES

Este balance nos dice que mayormente, los menores de edad fueron los que más presentaron el examen en los dos periodos

```
#mayores de edad y menores de edad comparados en los dos periodos
tipo_documento_2019 = pd.DataFrame(datos2019['estu_tipodocumento'].value_counts())
tipo_documento_2020 = pd.DataFrame(datos2020['estu_tipodocumento'].value_counts())
concat_tipo_documento_identidad = pd.concat([tipo_documento_2019, tipo_documento_2020], axis=1, ignore_index=True)
concat_tipo_documento_identidad = concat_tipo_documento_identidad.rename(columns={0: 'Tipo_doc_2019', 1: 'Tipo_doc_2020'})

ax = concat_tipo_documento_identidad.plot(kind='bar', stacked=False)
plt.xticks(rotation=0)
ax.legend(['2019-2', '2020-2'])
plt.show()
```

## Los 5 países con más estudiantes cursando el ICFES durante los dos periodos (Diferentes de Colombia)

Venezuela encabeza la lista, teniendo una subida abrupta de un año a otro, mientras que los siguientes en la lista, aunque presentaron una subida con diferencia a 2019-2 no representan mayor importancia a los datos

```
merged_estudiantes_por_pais = pd.concat([estudiantes_por_pais_2019, estudiantes_por_pais_2020], axis=1)
merged_estudiantes_por_pais = merged_estudiantes_por_pais['count'].fillna(0)


#Se añade fila que suma los estudiantes por nacionalidad en los dos periodos
merged_estudiantes_por_pais['Cantidad Total por país'] = merged_estudiantes_por_pais.sum(axis=1)
merged_estudiantes_por_pais = merged_estudiantes_por_pais.sort_values(by='Cantidad Total por país', ascending=False)
print(len(merged_estudiantes_por_pais['count']))

barras_estudiantes_por_pais = merged_estudiantes_por_pais.copy()
barras_estudiantes_por_pais = barras_estudiantes_por_pais.drop(columns=['Cantidad Total por país'])

ax = barras_estudiantes_por_pais[1:6].plot(kind='bar', stacked=False)
ax.legend(['2019-2', '2020-2'])
plt.show()
```
